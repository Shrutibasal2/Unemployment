# Data Visualization

## Univariate Analysis: It means analysing single variable through Data visualization

```python
plt.figure(figsize=(12,8))
sns.countplot(y=data.States)
plt.title('States')
plt.show()
#here we can see that their are only two states Jammu and Kashmir ,Sikkim which are present less as compared to other states
```

```python
colors = ['#4F6272', '#B7C3F3', '#DD7596', '#8EB897']
plt.figure(figsize=(12,8))
plt.pie(data['Region'].value_counts(),autopct='%.f%%',shadow=True,colors=colors,labels=['North','South','West','East','Northeast'])
plt.title('Regions')
plt.show()
```

```python
#Here we can infer that most of unemployment happen in North Region of every country after that most of region effect is South
```

```python
plt.figure(figsize=(12,7))
plt.hist(data['Estimated Unemployment Rate'],color='purple')
plt.text(x=45,y=20,s='skew='+' '+str(data['Estimated Unemployment Rate'].skew()))
plt.title('Estimated Unemployment Rate')
plt.show()
#here we can say that most of Unemployment Rate is consist between 0-10%
```

```python
plt.figure(figsize=(10,5))
sns.distplot(data["Estimated Labour Participation Rate"],color='r')
plt.show()
#Here we say that most of Estimated Labour Participation Rate is 40%
```
